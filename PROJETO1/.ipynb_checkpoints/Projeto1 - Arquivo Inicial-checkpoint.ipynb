# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
from time import sleep

# Passo a passo

# pyautogui.click -> clique com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas (ex: Ctrl + D)
# pyautogui.position -> indica a posição do mouse (x, y)
# pyautogui.PAUSE = seconds -> pausa entre comandos do pyautogui
# pyperclip.copy() -> copiar o texto

pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema da empresa(no link)
pyautogui.hotkey('ctrl', 't')
pyautogui.write(r'https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema')
pyautogui.press('enter')

# pode ser que o navegador tenha que carregar
sleep(3)

# Passo 2: Fazer login
# clicar no espaço de login
pyautogui.click(x=942, y=373)

# escrever login
pyautogui.write('felipe_login')

# clicar no espaço de senha
pyautogui.click(x=948, y=457)

# escrever senha
pyautogui.write('felipesenha123')

# clicar no botão acessar
pyautogui.click(x=971, y=527)

sleep(2)

# Passo 3: Exportar a base de dados
# clicar em cima do arquivo
pyautogui.click(x=380, y=416)

# clicar nos 3 pontinhos
pyautogui.click(x=1714, y=186)

# clicar em download
pyautogui.click(x=1501, y=594)
sleep(3)
```

```python
sleep(5)
print(pyautogui.position())
```

```python
# Passo 4: Calcular os indicadores
import pandas as pd

tabela = pd.read_csv(r'C:\Users\User\Desktop\INTENSIVAOpython\PROJETO1\Compras.csv', sep=';')
display(tabela)

total_gasto = tabela['ValorFinal'].sum()
qtd = tabela['Quantidade'].sum()
preco_medio = total_gasto / qtd

print(total_gasto)
print(qtd)
print(f'{preco_medio:.2f}')
```

```python
import pyperclip
# Passo 5: Enviar um e-mail para o meu chefe

pyautogui.PAUSE = 1

# entrar no email: https://mail.google.com/mail/u/0/#inbox
pyautogui.hotkey('ctrl', 't')
pyautogui.write(r'https://mail.google.com/mail/u/0/#inbox')
pyautogui.press('enter')

sleep(3)

# clicar no botão escrever
pyautogui.click(x=43, y=207)

# preencher as informações do email
# destinatário
pyautogui.write(r'pikatofly17@gmail.com')
pyautogui.press('tab') # escolher o destinatário

pyautogui.press('tab') # passar para o campo assunto
pyperclip.copy('Relatório de Compras') # copia o texto
pyautogui.hotkey('ctrl', 'v')

pyautogui.press('tab') # passando para o corpo do email
pyautogui.write('')


# enviar
# pyautogui.hotkey('ctrl', 'enter')
```
